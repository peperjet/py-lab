```python
# 루프를 이용한 반복 작업

# for 루프 : 정해진 범위나 시퀀스 순회할 때 사용
fruits = ["사과", "바나나", "체리"]
for fruit in fruits:
    print(f"내가 좋아하는 과일: {fruit}")

# range를 이용한 반복
for i in range(3):
  print(f"{i+1}번째 실행중 ...")
```

```python
while True: # 올바른 값을 입력할 때까지 무한 반복
    print("\n--- 등급 확인 프로그램 ---")
    user_input = input("점수를 입력하세요 (종료하려면 'q' 입력): ")

    # 종료 조건 처리
    if user_input.lower() == 'q':
        print("프로그램을 종료합니다.")
        break

    try:
        # 1. 형변환
        score = int(user_input)

        # 2. 범위 유효성 검사 (0~100점 사이인지)
        if score < 0 or score > 100:
            print("오류: 점수는 0점에서 100점 사이여야 합니다.")
            continue # 루프의 처음으로 돌아감

        # 3. 조건문을 통한 등급 판별
        if score >= 90:
            print(f"결과: {score}점은 A등급입니다.")
        elif score >= 80:
            print(f"결과: {score}점은 B등급입니다.")
        elif score >= 70:
            print(f"결과: {score}점은 C등급입니다.")
        else:
            print(f"결과: {score}점은 F등급입니다.")

    except ValueError:
        # 4. 예외 처리: 문자를 입력했을 경우
        print("오류: 숫자만 입력 가능합니다. (예: 85)")

print("이용해 주셔서 감사합니다!")
```
