```python
import sympy as sp
# 예제 2.

# 1. 사진 속 행렬 A를 그대로 입력합니다.
# 각 대괄호 [ ]는 행렬의 한 줄(Row)을 의미해요.
A = sp.Matrix([
    [1, 1, 1, 4, 4],
    [2, 3, 4, 9, 16],
    [-2, 0, 3, -7, 11]
])

print("--- [1단계] 우리가 처음 마주한 복잡한 행렬 ---")
display(A)

# 2. RREF(기약 행 사다리꼴)로 변환합니다.
# 이게 아까 우리가 1층 대장으로 2, 3층을 청소했던 그 과정의 결과물이에요.
rref_matrix, pivot_columns = A.rref()

print("\n--- [2단계] 모든 방해꾼(숫자)을 0으로 청소한 결과(RREF) ---")
display(rref_matrix)

print(f"\n--- [3단계] 정답 포인트(대장이 있는 위치): {pivot_columns} ---")
```

```python
# 예제 3
# 1. 왼쪽 숫자들(A)과 오른쪽 결과값(b)을 준비합니다.
A = sp.Matrix([
    [2, 4, 6],
    [2, -1, 1],
    [3, 0, -1]
])
b = sp.Matrix([18, 8, 3])

print("--- [준비] 범인들의 관계식 ---")
display(A, b)

# 2. '첨가 행렬' 만들기 (A 옆에 b를 딱 붙이는 거예요)
# 12살 버전: "용의자 명단 옆에 결과 점수를 붙여서 한 표로 만들자!"
Ab = A.row_join(b)

print("\n--- [1단계] 합체된 행렬 (첨가 행렬) ---")
display(Ab)

# 3. RREF로 청소하기
rref_matrix, _ = Ab.rref()

print("\n--- [2단계] 청소 끝! 범인이 누구인지 보이나요? ---")
display(rref_matrix)

# 결과를 리스트 형태로 예쁘게 출력하기
sol = rref_matrix[:, -1]
print(f"최종 정답은: x={sol[0]}, y={sol[1]}, z={sol[2]} 입니다!")
```

```python
# 예제 4번 데이터 입력
A = sp.Matrix([
    [1, -2, 1],
    [2, -2, 1],
    [3, 1, -5],
    [0, -1, 2],
    [-6, 0, 7]
])
b = sp.Matrix([7, 5, 0, -4, -10])

# 첨가 행렬 만들고 RREF 실행
Ab = A.row_join(b)
rref_result, _ = Ab.rref()

print("--- 예제 4번의 RREF 결과 ---")
display(rref_result)
```

```python
# 예제 5

# 1. 데이터 입력 (x1 ~ x6, 결과 b)
A = sp.Matrix([
    [1, 3, -2, 0, 2, 0],
    [2, 6, -5, -2, 4, -3],
    [0, 0, 5, 10, 0, 15],
    [2, 6, 0, 8, 4, 18]
])
b = sp.Matrix([0, -1, 5, 6])

# 2. 첨가 행렬(A와 b를 합치기) 및 RREF 계산
Ab = A.row_join(b)
rref_result, pivot_cols = Ab.rref()

print("--- 예제 5번의 RREF 결과 ---")
display(rref_result)

# 대장(1)이 있는 열의 번호도 함께 확인하면 해석이 더 쉬워요!
print(f"대장(Pivot)이 있는 열 번호: {pivot_cols}")
```

```python
# 행렬연산 합, 스칼라배
# SymPy
A = sp.Matrix([[1, 2, -4], [-2, 1, 3]])
B = sp.Matrix([[0, 1, 4], [-1, 3, 1]])
C = sp.Matrix([[1, 1], [2, 2]])

print("A + B =")
print(A + B)

print("\n2A =")
print(2*A)

print("\n(-1)C =")
print(-1*C)
```

```python
# --- [예제 1] 행렬의 합과 스칼라배 ---
# 똑같은 자리끼리 더하거나, 모든 숫자에 똑같이 곱하는 연산입니다.
A1 = sp.Matrix([[1, 2, -4], [-2, 1, 3]])
B1 = sp.Matrix([[0, 1, 4], [-1, 3, 1]])
C1 = sp.Matrix([[1, 1], [2, 2]])

print("=== 예제 1 결과 ===")
print("1) A + B (덧셈):")
display(A1 + B1)

print("\n2) 2A (2배 곱하기):")
display(2 * A1)

print("\n3) (-1)C (부호 바꾸기):")
display(-1 * C1)


# --- [예제 2] 행렬의 곱 (AB) ---
# 가로줄과 세로줄을 교차해서 곱하고 더하는 조금 더 복잡한 연산입니다.
A2 = sp.Matrix([[1, 2, -1], [3, 1, 0]])
B2 = sp.Matrix([[-2, 1], [0, -3], [2, 1]])

print("\n=== 예제 2 결과 ===")
print("A * B (행렬 곱셈):")
# SymPy에서는 그냥 * 기호를 쓰면 행렬 곱셈을 해줍니다!
display(A2 * B2)
```

```python
# --- [예제 3] 전치행렬 (Transpose) ---
A3 = sp.Matrix([[1, -2, 3], [4, 5, 0]])
B3 = sp.Matrix([[1, 2, -4], [3, -1, 2], [0, 5, 3]])

print("\n=== 예제 3: 전치행렬 (가로세로 뒤집기) ===")
print("A의 전치행렬 (A^T):")
display(A3.T)  # .T 또는 .transpose() 사용
print("B의 전치행렬 (B^T):")
display(B3.T)
```

```python
import random # random 모듈 추가

# --- [예제 4] 임의의 행렬 생성 및 전치 ---
# SymPy의 randMatrix는 minVal, maxVal 인수를 지원하지 않으므로, 직접 생성합니다.

rows, cols = 7, 12
min_val, max_val = -10, 10

matrix_elements = []
for _ in range(rows):
    row_elements = [random.randint(min_val, max_val) for _ in range(cols)]
    matrix_elements.append(row_elements)

A4 = sp.Matrix(matrix_elements)

print("\n=== 예제 4 결과 (임의의 7x12 행렬) ===")
print("A =")
display(A4)
print("A^T (뒤집힌 행렬) =")
display(A4.T)
```
